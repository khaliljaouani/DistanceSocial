```python
import cv2
import math
cap = cv2.VideoCapture("Distance1.avi")
body_classifer = cv2.CascadeClassifier("fullbody.xml")

while cap.isOpened():
    ret, frame = cap.read()
    bodys = body_classifer.detectMultiScale(frame)
    m="Projet Realiser Par JAOUANI/LOUKILI/ALLAF"
    cv2.putText(frame, m, (5, 550), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (0, 0, 255), 2)

    l = []
    lf = []
    i = 1
    for (x, y, w, h) in bodys:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)
        s = str(i)  #Numero de personne
        cv2.putText(frame, s, (x, y), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2) #inserer le numero de visage
        i += 1
        l = []
        l.append(x)
        l.append(y)
        lf.append(l)
        print(l)
    print(lf)
    distancePersonne = ""
    for i in range(len(lf)):
        for j in range(i + 1, len(lf)):
            d = math.sqrt(((lf[j][1] - lf[i][1]) ** 2) + ((lf[j][0] - lf[i][0]) ** 2))
            print("P", i + 1, "- P", j + 1, "=", d)
            if d < 100:
                distancePersonne += "Personne " + str(i + 1) + " et Personne " + str(j + 1) + " ; "
                cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 0, 255), 2)
    distancePersonne += " ne suivent pas la distanciation sociale "
    print(distancePersonne)
    if ret == True:
        cv2.imshow("Distance Sociale", frame)
        key = cv2.waitKey(40)
        if key == 27:
            break
    else:
        break
cv2.destroyAllWindows()
cap.release()
```
